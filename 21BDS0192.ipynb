```r
install.packages("moments")
install.packages("mice")
install.packages("ggplot2")
install.packages("dplyr")

# RAGHAV AGARWAL
# 21BDS0192

# Load necessary libraries
library(dplyr)
library(readr)
library(moments)  # For skewness and kurtosis
library(ggplot2)
library(mice)  # For multiple imputation methods

# Read the dataset from the given URL
data <- read_csv("https://raw.githubusercontent.com/salemprakash/EDA/main/Data/catholic.csv")

# Display the first few rows of the dataset
head(data)

# ------------------------------------
# 1. Data Transformation
# ------------------------------------
normalize_range <- function(x, new_min, new_max) {
  old_min <- min(x, na.rm = TRUE)
  old_max <- max(x, na.rm = TRUE)
  (x - old_min) / (old_max - old_min) * (new_max - new_min) + new_min
}

# Transform data to different ranges
data$transformed_1 <- normalize_range(data[[2]], -3, 3)  # [-3,3]
data$transformed_2 <- normalize_range(data[[2]], -1, 1)  # [-1,1]
data$transformed_3 <- normalize_range(data[[2]], 0, 1)   # [0,1]

# ------------------------------------
# 2. Summary Statistics
# ------------------------------------
summary_stats <- function(x) {
  return(list(
    mean = mean(x, na.rm = TRUE),
    median = median(x, na.rm = TRUE),
    mode = as.numeric(names(sort(table(x), decreasing = TRUE)[1])),
    variance = var(x, na.rm = TRUE),
    sd = sd(x, na.rm = TRUE)
  ))
}

# Apply summary statistics to the first numeric column
data_summary <- summary_stats(data[[2]])
print(data_summary)

# Read X, Y values and compute covariance & correlation
X <- data[[2]]
Y <- data[[3]]
cov_XY <- cov(X, Y, use = "complete.obs")
corr_XY <- cor(X, Y, use = "complete.obs")

print(paste("Covariance:", cov_XY))
print(paste("Correlation:", corr_XY))

# Outlier detection using IQR method
outlier_detection <- function(x) {
  Q1 <- quantile(x, 0.25, na.rm = TRUE)
  Q3 <- quantile(x, 0.75, na.rm = TRUE)
  IQR <- Q3 - Q1
  lower_bound <- Q1 - 1.5 * IQR
  upper_bound <- Q3 + 1.5 * IQR
  return(which(x < lower_bound | x > upper_bound))
}

outliers <- outlier_detection(data[[2]])
print("Outliers detected at:")
print(outliers)

# ------------------------------------
# 3. Estimation of Missing Data
# ------------------------------------
# Global method: Replace missing values with mean
fill_with_mean <- function(x) {
  x[is.na(x)] <- mean(x, na.rm = TRUE)
  return(x)
}
data$filled_global <- fill_with_mean(data[[2]])

# Class-based method: Replace missing values with median
fill_with_median <- function(x) {
  x[is.na(x)] <- median(x, na.rm = TRUE)
  return(x)
}
data$filled_class_based <- fill_with_median(data[[2]])

# Multiple Imputation
imputed_data <- mice(data, method = "pmm", m = 5, maxit = 5, seed = 123)
data_imputed <- complete(imputed_data)

# Display the transformed and imputed data
head(data)
```
